```python
import os
import sys
from pathlib import Path

# Get current directory
cwd = Path.cwd()
print(f"Current Working Directory: {cwd}")

# Look for README to find root
root = None
current = cwd
for _ in range(5):
    if (current / "README.md").exists():
        root = current
        break
    current = current.parent

if root:
    print(f"Project Root Found: {root}")
    src_path = root / "src"
    utils_path = src_path / "utils"
    
    print(f"Checking for 'src' folder: {'EXISTS' if src_path.exists() else 'MISSING'}")
    if src_path.exists():
        print(f"Checking for 'src/__init__.py': {'EXISTS' if (src_path / '__init__.py').exists() else 'MISSING'}")
        print(f"Checking for 'src/utils/__init__.py': {'EXISTS' if (utils_path / '__init__.py').exists() else 'MISSING'}")
        print(f"Checking for 'src/utils/config.py': {'EXISTS' if (utils_path / 'config.py').exists() else 'MISSING'}")
else:
    print("Could not find README.md in this or parent directories.")

print(f"\nFirst 3 sys.path entries:\n{sys.path[:3]}")
```
