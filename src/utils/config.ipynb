```python
from pathlib import Path
import sys

# Detect project root directory
def get_repo_root():
   

    # If running in Google Colab
    if 'google.colab' in sys.modules:
        print("Running in Google Colab environment.")
        return Path('/content')

    # Running locally
    else:
        print("Running in local environment.")

        # Start from current directory
        current = Path().resolve()

        # Walk upward until we find project root (identified by README.md)
        while current != current.parent:
            if (current / "README.md").exists():
                return current
            current = current.parent

        raise Exception("Project root not found. Make sure README.md exists.")


# Store root once
REPO_ROOT = get_repo_root()


# Generic folder getter
# Flexible function to build paths dynamically
def get_path(*folders):
    
    return REPO_ROOT.joinpath(*folders)
```
