```python
# anonymize the data by removing PI information from text columns

import os
import pandas as pd
import re

# Folder containing Excel files
folder_path = r"C:\Users\LENOVO\OneDrive\Documents\Strath\Masters\sentiment-analysis-tool\Data\Raw Data"

# Regex patterns for common PII
patterns = {
    'EMAIL': r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b',
    'PHONE': r'\+?\d[\d\s.-]{9,}\d',
    'ID': r'\b\d{7,}\b',
    'DATE': r'\b\d{1,2}[/-]\d{1,2}[/-]\d{2,4}\b',  # 01/01/2000 or 1-1-20
    'CREDIT_CARD': r'\b\d{4}[\s-]?\d{4}[\s-]?\d{4}[\s-]?\d{4}\b',
    'IP_ADDRESS': r'\b(?:\d{1,3}\.){3}\d{1,3}\b',
}

# Function to anonymize text using regex
def anonymize_text_regex(text):
    text = str(text)
    for key, pattern in patterns.items():
        text = re.sub(pattern, f"<{key}>", text)
    return text

# Process all Excel files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith(".xlsx"):
        file_path = os.path.join(folder_path, file_name)
        print(f"Processing {file_name}...")

        # Read Excel file
        df = pd.read_excel(file_path)

        # Automatically detect text columns
        text_columns = df.select_dtypes(include='object').columns.tolist()

        # Apply regex anonymization to each text column
        for col in text_columns:
            matches_count = 0
            for key, pattern in patterns.items():
                matches_count += df[col].astype(str).str.count(pattern).sum()
            if matches_count > 0:
                print(f"Found {matches_count} PII entries in column '{col}'")
            df[col] = df[col].apply(anonymize_text_regex)

        # Save anonymized copy
        anonymized_file = os.path.join(folder_path, f"anonymized_{file_name}")
        df.to_excel(anonymized_file, index=False)
        print(f"Saved anonymized file: {anonymized_file}")

print("All files anonymized successfully!")
```
